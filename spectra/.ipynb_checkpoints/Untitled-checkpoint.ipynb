```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import datetime
from astropy.time import Time
import os
```

```python
cwin = {'fex6547':(6480,6640),
       'fevii5159':(5280,5390)}
lwin = {'fex6547':(6560,6625),
       'fevii5159':(5320,5370)}
```

```python
def calc_pseudo_lineflux(lam,flux,cwin,lwin,
                         lrng=10,fig=False):
    '''calculate pseudo-continuum, find line flux'''
    # pseudo-continuum
    ind_lowc = np.argmin(np.abs(lam-cwin[0]))
    ind_highc = np.argmin(np.abs(lam-cwin[1]))
    med_low = np.median(flux[ind_lowc-lrng:ind_lowc+lrng])
    med_high = np.median(flux[ind_highc-lrng:ind_highc+lrng])
    a = (med_high-med_low)/(cwin[1]-cwin[0])
    b = med_low
    # line flux
    ind_lowl = np.argmin(np.abs(lam-lwin[0]))
    ind_highl = np.argmin(np.abs(lam-lwin[1]))
    line_flux = 0
    for ii in range(ind_lowl,ind_highl):
        binw = 0.5*(lam[ii+1]-lam[ii-1])
        fd = flux[ii]
        pc = a*(lam[ii]-cwin[0])+b
        line_flux += binw*(fd-pc)
    if fig:
        fig,ax = plt.subplots(1,figsize=(5,5))
        ax.plot(lam[ind_lowc-20:ind_highc+20],flux[ind_lowc-20:ind_highc+20],c='b')
        ax.plot( (lam[ind_lowc],lam[ind_highc]),(b,b+a*(cwin[1]-cwin[0])),c='r' )
        ax.vlines(lwin[0],0,1e-14)
        ax.vlines(lwin[1],0,1e-14)
        # find rough estimate (appr. as triangle)
        fmaxl = np.amax(flux[ind_lowl:ind_highl])
        fminl = 0.5*(med_low+med_high)
        ax.hlines(fmaxl,cwin[0],cwin[1],ls='--',color='g')
        ax.hlines(fminl,cwin[0],cwin[1],ls='--',color='g')
        ax.set_ylim(0.9*med_low,1.05*np.amax(flux[ind_lowc-20:ind_highc+20]))
        print('bin width =',binw)
        print('rough estimate (triangle) =',0.5*(lwin[1]-lwin[0])*(fmaxl-fminl))
        plt.show()
    return (line_flux)

def find_p_from_t (arrin1,arrin2):
    r = stats.pearsonr(arrin1,arrin2)
    pcc = r[0]
    bn = len(arrin1)
    Tpcc = pcc*np.sqrt((bn-2)/(1-pcc*pcc))
    pval = stats.t.sf(np.absolute(Tpcc), bn-2)
    #print 'Pearson', r,pval
    return pval*2.

def find_rs (arrin1,arrin2): #p-value for t-statistic derived from Spearman correlation coefficient.
    r1 = stats.rankdata(arrin1)
    r2 = stats.rankdata(arrin2)
    bn = len(arrin1)
    Spcc = 1 - 6./(bn*(bn*bn-1))*sum(np.power(r2-r1,2))
    #print 'Spearman', stat.spearmanr(arrin1,arrin2),Spcc
    return Spcc
```

```python
# Calculate line fluxes for FeX & FeVII
greyshift = np.loadtxt('../../FAST_greyshifts').T
op = []
for f in os.listdir('.'):
    if f.endswith('.spec'):
        print(f)
        sp = np.loadtxt(f).T
        mjd = Time(datetime.datetime.strptime(f.split('.')[0][-8:],'%Y%m%d'),format='datetime').mjd
        if int(mjd) in greyshift[0]:
            gshift = greyshift[1][np.argwhere(greyshift[0]==int(mjd))]
            sp[1] = sp[1]/gshift
            print(mjd,'greyshifted!')
        lf10 = calc_pseudo_lineflux(sp[0],sp[1],cwin['fex6547'],lwin['fex6547'],fig=False)
        lf7 = calc_pseudo_lineflux(sp[0],sp[1],cwin['fevii5159'],lwin['fevii5159'],fig=False)
        op.append((mjd,lf7,lf10))
op=np.array(op)
```

```python
# Combine results with HeII fluxes
dfhe = pd.read_csv('../HeII_fluxes',delim_whitespace=True)
dffe = pd.DataFrame(data=op,columns=['date','fex','fevii'])
df = pd.merge(dfhe,dffe,how='inner',on=['date'])
df.drop(df.index[df.fevii>3e-14],inplace=True)
df.drop(df.index[df.fex<0],inplace=True)
# normalise fluxes
nind = 12
df['amp_b_norm']=df.amp_b/df.iloc[nind].amp_b
df['amp_b_norm_e']=df.amp_b_e/df.iloc[nind].amp_b
df['fex_norm'] = df.fex/df.iloc[nind].fex
df['fevii_norm'] = df.fevii/df.iloc[nind].fevii
df
```

```python
fig,axs = plt.subplots(1,2,figsize=(12,6))
# plot 1
x,y = df.amp_b*1e14,df.fex*1e14
axs[0].scatter(x,y,c='r')
lg = stats.linregress(x, y)
axs[0].plot(x, lg.intercept + lg.slope*x, 'k', label='Fitted line')
rs = find_rs(x,y)
pval = find_p_from_t(x,y)
axs[0].text(0.1,0.75,
           r'$r_s$='+'{:.2}'.format(rs)+'\np ='+'{:.2}'.format(pval),
           transform=axs[0].transAxes,fontsize=16,
           color='grey',bbox=dict(facecolor='w',edgecolor='w'))
# plot 2
x,y = df.amp_b*1e14,df.fevii*1e14
axs[1].scatter(x,y,c='b')
lg = stats.linregress(x, y)
axs[1].plot(x, lg.intercept + lg.slope*x, 'k', label='Fitted line')
rs = find_rs(x,y)
pval = find_p_from_t(x,y)
axs[1].text(0.1,0.75,
           r'$r_s$='+'{:.2}'.format(rs)+'\np ='+'{:.2}'.format(pval),
           transform=axs[1].transAxes,fontsize=16,
           color='grey',bbox=dict(facecolor='w',edgecolor='w'))
plt.show()
```
